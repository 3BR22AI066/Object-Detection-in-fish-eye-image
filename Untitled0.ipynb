```python
from ultralytics import YOLO
```

```python
import os

labels_dir = "/content/drive/MyDrive/Fisheye8K_all_including_train&test/train/labels"

classes = set()
for file in os.listdir(labels_dir):
    if file.endswith(".txt"):
        with open(os.path.join(labels_dir, file), "r") as f:
            for line in f:
                if line.strip():
                    cls_id = line.split()[0]
                    classes.add(int(cls_id))

print("Found class IDs:", sorted(list(classes)))
print("Total classes:", len(classes))
```

```python
import json

with open("/content/drive/MyDrive/Fisheye8K_all_including_train&test/train/train.json", "r") as f:
    data = json.load(f)

categories = data["categories"]
class_names = [c["name"] for c in categories]

print("Class names:", class_names)
print("Total classes:", len(class_names))
```

```python
dataset_yaml = """
train: /content/drive/MyDrive/Fisheye8K_all_including_train&test/train/images
val: /content/drive/MyDrive/Fisheye8K_all_including_train&test/test/images

nc: 5
names: ['Bus', 'Bike', 'Car', 'Pedestrian', 'Truck']
"""

with open("fisheye_dataset.yaml", "w") as f:
    f.write(dataset_yaml)

!cat fisheye_dataset.yaml
```

```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')  # load pretrained model

model.train(
    data='fisheye_dataset.yaml',  # your dataset config
    epochs=50,                    # number of training epochs
    imgsz=640,                    # image size
    batch=16,                     # adjust depending on GPU memory
    name='fisheye_yolov8',        # experiment name
    workers=4                     # number of data loader workers
)
```

```python
from IPython.display import Image, display
display(Image(filename='runs/detect/fisheye_yolov8/results.png'))
```

```python
results = model.predict(
    source='/content/drive/MyDrive/Fisheye8K_all_including_train&test/test/images',
    stream=True,
    conf=0.4,
    save=True
)
```

```python
from ultralytics import YOLO

# Load your trained model
model = YOLO("/content/runs/detect/fisheye_yolov8/weights/best.pt")

# Export to TensorFlow Lite (for mobile)
model.export(format="tflite", int8=True)
```

```python
import tensorflow as tf

# Correct absolute path
model_path = "/content/runs/detect/fisheye_yolov8/weights/best_saved_model/best_int8.tflite"

interpreter = tf.lite.Interpreter(model_path=model_path)
interpreter.allocate_tensors()

print("âœ… TFLite model loaded successfully!")
```

```python
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print("Input details:", input_details)
print("Output details:", output_details)
```
